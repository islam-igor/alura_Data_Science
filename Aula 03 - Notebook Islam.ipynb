```python
#Preço do metro quadrado por bairro e plotar em um gráfico ideal

dados_vendas.head()

dados_vendas['Valor_anuncio_float']/dados_vendas['Metragem'] #valor por metro quadrado de cada imóvel
#criar uma coluna com esses valores
dados_vendas['Valor_m2'] = dados_vendas['Valor_anuncio_float']/dados_vendas['Metragem'] 
dados_vendas.sample(5)
```

```python
dados_vendas.groupby('Bairro').mean()    #agrupa os Bairros fazendo a média das colunas com números
```

```python
dados_vendas.groupby('Bairro').sum()

dados_bairro = dados_vendas.groupby('Bairro').sum()
dados_bairro
```

```python
dados_bairro["Valor_m2_bairro"] = dados_bairro["Valor_anuncio_float"]/dados_bairro["Metragem"]
dados_bairro
```

```python
dados_bairro.reset_index() #transforma Bairro em Index novamente, mas nao salva a alteração
dados_bairro.reset_index(inplace= True) #retorna ao proprio dado a substituição
dados_bairro
```

```python
dados_vendas['Bairro'].value_counts() #contar as vezes que aparece cada bairro
dados_vendas['Bairro'].value_counts().index #aparece o nome dos bairros na ordem decrescente 
top_bairros = dados_vendas['Bairro'].value_counts()[:10].index #filtrando os dois primeiros (os maiores)
```

```python
dados_bairro.query("Bairro in @top_bairros") #.query faz uma filtragem
```

```python
plt.figure(figsize=(15,8))
ax = sns.barplot(x= 'Bairro', y = 'Valor_m2_bairro', data=dados_bairro.query("Bairro in @top_bairros"))
plt.show()
```

```python
plt.figure(figsize=(15,8))
ax = sns.barplot(x= 'Bairro', y = 'Valor_m2_bairro', data=dados_bairro.query("Bairro in @top_bairros"))
ax.tick_params(axis= 'x', rotation = 45)
plt.show()
```

```python
plt.figure(figsize=(15,8))
ax = sns.boxplot(data = dados_vendas.query('Bairro in @top_bairros'), x = 'Bairro', y = 'Valor_anuncio_float')
ax.tick_params(axis= 'x', rotation = 45)
plt.show()
```

```python
plt.figure(figsize=(15,8))
ax = sns.boxplot(data = dados_vendas.query('Bairro in @top_bairros and Metragem < 30000'), x = 'Bairro', y = 'Metragem')
ax.tick_params(axis= 'x', rotation = 45)
plt.show()
```
