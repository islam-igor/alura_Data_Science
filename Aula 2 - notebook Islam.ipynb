```python
dados[['Moeda', 'Valor_anuncio', 'Tipo_anuncio']] = dados['Valor'].str.split(expand = True)
dados.sample(10)
```

```python
#eliminar tipos de anuncio que não são de vendas, valores de aluguel /mes, /ano
#para isso, vamos permanecer com os dados 'None' (vazios)

dados['Tipo_anuncio'].isnull() #transforma em True e False (True quando for vazio)

#colocando --> dados['Tipo_anuncio'].isnull() <-- filtraremos nossos dados para mostrarem apenas os True
dados[dados['Tipo_anuncio'].isnull()]


```

```python
dados[dados['Tipo_anuncio'].isnull()]['Tipo_anuncio'].unique() 
#confirmando se temos apenas os valores Vazios
```

Agora temos que salvar o nosso DataFrame filtrado , para que nele só haja as informações de vendas

Mas para nao haver perdas de informações do dataFrame original 'dados'

iremos criar uma nova variável, e trabalharemos nela de agora por diante.


```python
dados_vendas = dados[dados['Tipo_anuncio'].isnull()]
dados_vendas
```

```python
dados['Valor_anuncio'].str.replace('.','') #trocou os pontos '.' por nada/vazio ''
```

```python
#covnertendo Valor_anuncio que antes era Object/string para Float
# --> dados_vendas['Valor_anuncio'].str.astype(float)  <--- o codigo nao roda porque não ficou salvo na memoria o replace.

#entao tem que fazer a conversão no mesmo código que faz o replace.
dados['Valor_anuncio'].str.replace('.','').astype(float)
```

```python
#é necessário gravar a informação. seja criando uma nova coluna, uma nova variável , ou substituindo a dados_vendas

dados_vendas['Valor_anuncio_float'] = dados_vendas['Valor_anuncio'].str.replace('.','').astype(float)
dados_vendas
```

```python
dados_vendas['Valor_anuncio_float'].plot.hist()
```

```python
# o grafico ficou ruim de ter insight, procuramos formas de melhorar
dados_vendas['Valor_anuncio_float'].plot.hist(bins = 50)
#bins aumenta o número de barras
```
