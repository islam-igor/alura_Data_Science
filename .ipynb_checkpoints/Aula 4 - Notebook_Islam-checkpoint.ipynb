```python
enderecos.head()
```

 no momento temos 3 bases.
 - base inicial de vendas
 - base do ibge com setores censitários
 - base de endereços 
 
 a base de endereços vai servir pra vincular a base vendas com base de setor censitário

```python
#filtrando os endereços apenas do estado de São Paulo
enderecos_sp = enderecos.query("sigla_uf == 'SP'")
enderecos_sp 
```

```python
#Criando uma variável chamada rua no qual vai constar o nome do logradouro + nome do endereço
#pois na base em questão elas sao separadas, e a nossa de vendas são juntas.
enderecos_sp["rua"] = enderecos_sp["tipo_logr"] + " " + enderecos_sp["logr_nome"]

#padronizando o nome para que fique todo em minúsculo e que não tenham espaços indesejáveis(strip)
enderecos_sp["rua"] = enderecos_sp["rua"].str.lower().str.strip()
enderecos_sp.head()
```

```python
dados_vendas
#no banco de dados de dados_vendas o nome da Rua tem o número da Rua e tem algumas letras maiúsculas, teremos que padronizar.
```

```python
dados_vendas["Rua"].str.extract(r'(^[\w ]+)')[:10] #código complexo, recomendado ler o artigo :https://medium.com/creditas-tech/incrementando-dados-geogr%C3%A1ficos-com-o-censo-nacional-do-ibge-54d342c4bdcf

# o codigo em questão remove caracteres especiais: virgula, pontos e coisas do tipo. e pegando só o nome da rua, sem número
```

```python
#aplicando o extract e criando uma nova coluna na base de dados_venda
dados_vendas["apenas_rua"] = dados_vendas["Rua"].str.extract(r'(^[\w ]+)')
#padronizando todos os nomes em minusculo e sem espaços indesejáveis
dados_vendas["apenas_rua"] = dados_vendas["apenas_rua"].str.lower().str.strip()
dados_vendas.head()
```

```python
#agora precisamos vincular as duas bases, abstraindo apenas as colunas de interesse
#queremo Cep, Lat, Long da base de endereco_sp e dar um MERGE na base de dados_vendas.
dados_geo = pd.merge(left = dados_vendas, right = enderecos_sp[["rua", "cep", "latitude", "longitude"]], how = "left", left_on = "apenas_rua", right_on = "rua")
#esse merge criou um banco de dados com 90mil linhas porque tem endereços que vão ter lat/long diferentes, mesmo que proximos
#e para cada lat/long haverá uma nova linha, como isso pode ser problemático, deveremos filtrar
dados_geo
```

```python
#meu interesse é manter uma única linha para cada item dos meus dados_vendas, para isso adiciona-se uma extensão para
#remover as duplicatas das colunas dados vendas
#a query com cep>0 é porque havera alguns dados que nao vão ser encontradas na base de dados endereços

dados_geo = pd.merge(left = dados_vendas, right = enderecos_sp[["rua", "cep", "latitude", "longitude"]], how = "left", left_on = "apenas_rua", right_on = "rua").drop_duplicates(subset=dados_vendas.columns).query("cep > 0")
dados_geo 

#isso acontece porque nao estamos trabalhando com os numeros das ruas, ou seja, duas casas com numeros diferentes em uma
#mesma rua, tem lat e long diferentes, e possivelmente podem ter até ceps diferentes.
```

```python
ibge_sp.info()
#nao tem as ionformações de latitude e longitude, mas ela tem o código setor
#existe um outro banco de dados que vincular o codigo setor com as latitudes e longitudes do poligono referente aquele setor
#precisaremos buscá-la para integrar os bancos de dados do ibge_sp com dados_vendas
```

```python
import geopandas as gpd
```

```python
setor_censo = gpd.read_file("35SEE250GC_SIR.shp")
setor_censo.head()
#nota-se que CD_GEOCODI é o código sensitário

# e no final da tabela tempo o geometry com o polygono do setor
```

```python
#criando um setor censitário filtrado para o município  de são paulo
setor_censo_sp = setor_censo[setor_censo.NM_MUNICIP == "SÃO PAULO"]
#buscando o setor censitario que o ponto de lat/long determinado esta
setor_censo_sp[setor_censo_sp.contains(Point(-46.63, -23.58))]
```

```python
#criando um Point pra cada lat/long que sera informada pela base dados_geo(nossa base que foi merge (vendas e endereços))

dados_geo["Point"] = ""
for i in dados_geo.index:
    dados_geo["Point"][i] = Point(dados_geo["longitude"][i], dados_geo["latitude"][i])



```

```python
# pegando cada um dos meus pontos e procurando dentro do setor censitário_sp 
#se tem ou nao setor censitário dentro deaquele polygono e retornando o Codigo censitário
#criando uma nova coluna setor_censo para armazenar a qual censo aquele endereço pertence
# na qual essa coluna vai ter o mesmo codigo censitário do banco de dados do setor_censo_sp que contem os Poligonos
dados_geo['setor_censo'] = dados_geo["Point"].map(
    lambda x: setor_censo_sp.loc[setor_censo_sp.contains(x), 'CD_GEOCODI'].values
).str[0]
dados_geo

#esse código é muito demorado, por conta que ele vai analisar muitos pontos em diversos caminhos, portanto, 
#foi criado um arquivo csv dados_geo para ser importado de forma mais rápida

#dados_geo.to_csv("nome-do-arquivo.csv", index = False)
```

```python
#agora faz um merge unindo as bases dados_geo (contendo o POINT , Lat, Long, CEP) 
#e o banco de dados do IBGE que contem os POLIGONOS, utilzando a coluna setor censitário que é igual em ambas

dados_vendas_censo = pd.merge(left = dados_geo, right = ibge_sp, how = "left", left_on = "setor_censo", right_on = "Cod_setor")
dados_vendas_censo
```
